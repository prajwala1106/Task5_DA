```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('gender_submission.csv')
```

```python
# Display first few rows
print(df.head())
```

```python
# Basic info
print(df.info())
```

```python
# Summary stats
print(df.describe())
```

```python
# Value counts for Survived
print(df['Survived'].value_counts())
```

```python
sns.countplot(x='Survived', data=df, palette='Set2')
plt.title("Distribution of Survival")
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.ylabel("Count")
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%', colors=['salmon', 'skyblue'])
plt.title("Survival Rate")
plt.show()
```

```python
# Scatterplot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Survived', y='PassengerId', data=df, color='purple', s=50)
plt.title("Passenger ID vs Survival")
plt.xlabel("Passenger ID")
plt.ylabel("Survived (0 = No, 1 = Yes)")
plt.grid(True)
plt.show()
```

```python
# Compute correlation
corr = df.corr()

# Heatmap
plt.figure(figsize=(5, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
